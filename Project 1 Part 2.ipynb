 ### Primary Sources of Awareness

```python
import matplotlib.pyplot as plt
import seaborn as sns 

# Columns related to sources of awareness
source_columns = [
    'heard_from_ccm_web_site',
    'heard_from_social_media',
    'heard_from_community_event',
    'heard_from_family_member_or_friend',
    'heard_from_current_ccm_student',
    'heard_from_ccm_alumni',
    'heard_from_high_school_teacher',
    'heard_from_high_school_counselor'
]

# Count the occurrences of each source
source_counts = {}
for col in source_columns:
    source_counts[col.split('heard_from_')[-1].replace('_', ' ')] = data[col].value_counts().get('Yes', 0)

# Plotting
plt.figure(figsize=(14, 7))
sorted_sources = dict(sorted(source_counts.items(), key=lambda item: item[1], reverse=True))
sns.barplot(x=list(sorted_sources.keys()), y=list(sorted_sources.values()), palette="viridis")
plt.title('Primary Sources of Awareness')
plt.xlabel('Source')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The bar chart depicts the primary sources through which respondents became aware of CCM. Evidently, the "Family member or friend" and "High school counselor" are dominant sources of information, followed closely by "High school teacher" and "Current ccm student." CCM might want to improve its online presence and reach out to Morris County community event organizers to boost its presence in the community. Additionally, CCM can develop a plan to incentivize alumni to refer their family and friends to the college.

### Common Reasons for Interest in Computing

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top reasons for interest in computing
top_reasons = data['reason_interest_in_computing'].value_counts().head(10)

plt.figure(figsize=(12, 8))
top_reasons.plot(kind='barh', color=sns.color_palette("viridis", 10))
plt.title('Top 10 Reasons for Interest in Computing', fontsize=16)
plt.xlabel('Number of Respondents', fontsize=14)
plt.ylabel('Reason', fontsize=14)
plt.gca().invert_yaxis()  # Invert the y-axis for better readability
plt.tight_layout()

plt.show()
```

The predominant reason is "I would like to learn about computing."
Other notable reasons include interests in specific computing areas, job opportunities, and personal growth. CCM needs to expose more students to programming and help them build cool projects to share with their friends and family; in return, it might help with recruitment.

### Distribution of Respondents by Race/Ethnicity

```python
# Plotting the distribution of respondents by race/ethnicity
plt.figure(figsize=(12, 7))
sns.countplot(data=data, y='race_ethnicity', order=data['race_ethnicity'].value_counts().index, palette="viridis")
plt.title('Distribution of Respondents by Race/Ethnicity', fontsize=16)
plt.ylabel('Race/Ethnicity', fontsize=14)
plt.xlabel('Number of Respondents', fontsize=14)
plt.tight_layout()

plt.show()
```

The majority of respondents identify as "White/Caucasian."
Other racial/ethnic categories have comparatively fewer respondents. A significant effort to reach minority communities might be necessary because the data shows that they are severely underrepresented in this dataset. 

### Enrollment Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set_style("whitegrid")

# Enrollment Distribution
plt.figure(figsize=(12, 6))
course_counts = data['course_enrolled'].value_counts()
sns.barplot(x=course_counts.index, y=course_counts.values, palette="viridis")
plt.title('Enrollment Distribution')
plt.xlabel('Course Enrolled')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The bar chart above provides insights into the enrollment distribution among the respondents. It appears that "Computer Science" has the highest enrollment, followed by other courses. This suggests that Computer Science is a popular choice among the survey participants; therefore, CCM might want to find out why enrollment for CS is higher and adopt this blueprint to improve recruiting.
